```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
matches = pd.read_csv('ipldata/IPL Matches 2008-2020.csv')
deliveries = pd.read_csv('ipldata/IPL Ball-by-Ball 2008-2020.csv')
```

```python
matches.head()
```

```python
matches.shape
```

```python
deliveries.shape
```

```python
pd.value_counts(matches['winner'])
```

```python
# pd.value_counts(matches['winner'])
plt.hist(pd.value_counts(matches['winner']))
plt.show()
```

```python
plt.title("Winners Count")
plt.hist(pd.value_counts(matches['winner']), width=8)
plt.show()
```

```python
labels = pd.value_counts(matches['winner']).index
bars = pd.value_counts(matches['winner'])

plt.style.use('ggplot')
plt.figure(figsize=(12,6))
plt.title("Winners Count")
plt.bar(labels, bars, color='red')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Winners Count")

sns.countplot(data=matches, x=matches['winner'])

plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Winners Count")

sns.countplot(data=matches, x=matches['city'])

plt.xticks(rotation='vertical')
plt.show()
```

```python
toss_decision = pd.value_counts(matches['toss_decision'])
```

```python
toss_decision
```

```python
labels = toss_decision.index
```

```python
values = toss_decision.values
```

```python
plt.figure(figsize=(12,6))
plt.title("Toss decision")

plt.pie(values, labels=labels, startangle=90,counterclock=False,
       shadow=True, autopct='%1.2f%%', colors=['yellow','cyan'],
       explode=(0,0.2))
plt.legend(loc=1)
plt.show()
```

```python
matches['date'] = pd.to_datetime(matches['date'])
```

```python
matches['season'] = pd.to_datetime(matches['date']).dt.year
```

```python
matches.head()
```

```python
pd.value_counts(matches['season'])
```

```python
plt.figure(figsize=(12,6))
plt.title("Matches Per Season")

sns.countplot(data=matches, x=matches['season'])

plt.xticks(rotation='vertical')
plt.show()
```

```python
matches.drop_duplicates('season', keep='last')[['season','player_of_match','winner']]
```

```python
matches.drop_duplicates('season', keep='last')[['season','player_of_match','winner']].reset_index(drop=True)
```

```python
plt.figure(figsize=(12,6))
plt.title("Toss decision per season")

sns.countplot(data=matches, x=matches['season'], hue=matches['toss_decision'])

plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(12,6))
pd.value_counts(matches['season']).plot.bar(width=0.9, color=sns.color_palette('rocket'))
```

```python
plt.figure(figsize=(12,6))
deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).plot.bar(width=0.9,
                                                                                                   color=sns.color_palette('rocket'))
```

```python
plt.figure(figsize=(12,6))
deliveries.groupby(['batsman','id'])['batsman_runs'].sum().sort_values(ascending=False).head(10).plot.bar(width=0.9,
                                                                                                   color=sns.color_palette('rocket'))
```
