```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('headbrain.csv')
```

```python
X = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values
```

```python
X = X.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
y = sc.fit_transform(y)
```

```python
def cost_function(x,y,slope,inter):
    n = len(x)
    error = (sum(((slope * x + inter) - y) ** 2)) / n
    return error
```

```python
cost_function(X,y,0,0)
```

```python
def gradient_descent(x, y, epochs, alpha):
    n = len(x)
    cost_history = [0] * epochs
    slope = 0
    inter = 0
    for i in range(epochs):
        pred = slope * x + inter
        error = pred - y
        grad_slope = (2/n) * error.T.dot(x)
        #grad_slope = (2/n) * np.sum(error * x)
        grad_inter = (2/n) * np.sum(error)
        slope = slope - grad_slope * alpha
        inter = inter - grad_inter * alpha
        cost_history[i] = cost_function(x,y,slope,inter)
        #print(cost_history[i], slope, inter)
    return slope, inter, cost_history
```

```python
epochs = 200
alpha = 0.01
slope, inter, error = gradient_descent(X,y,epochs,alpha)
```

```python
predictions = []
for i in range(len(X)):
    y_pred = X[i] * slope[0][0] + inter
    predictions.append(y_pred)
```
