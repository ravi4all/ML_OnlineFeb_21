```python
while True:
    cv2.imshow('result', img)
    #27 - ASCII of escape
    if cv2.waitKey(2) == 27:
        break

cv2.destroyAllWindows()
```

```python
while True:
    cv2.rectangle(img, (10,10), (200,200), (0,0,0), 5)
    cv2.imshow('result', img)
    if cv2.waitKey(2) == 27:
        break

cv2.destroyAllWindows()
```

```python
img = cv2.imread('image_1.jpg')

img_1 = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
img_2 = cv2.resize(img, (1000,800))
stretch = cv2.resize(img, (900, 668), interpolation=cv2.INTER_LINEAR)

images = [img, img_1, img_2, stretch]
labels = ['orignal', 'small', 'large', 'stretch']

# for i in range(4):
#     plt.subplot(2,2,i+1)
#     plt.title(labels[i])
#     plt.imshow(images[i])

for i in range(4):
    while True:
        cv2.imshow(labels[i], images[i])
        if cv2.waitKey(2) == 27:
            break

    cv2.destroyAllWindows()
```
